## Background and Prior Work

The Pima Indians, also known as the Akimel O’odham, are a Native American tribe predominantly residing in the Salt River Pima–Maricopa Indian Community in Arizona, USA. This community has received attention in the realm of public health due to a high prevalence of diabetes, both Type-I and Type-II, among its members. Researchers have long been intrigued by the unusually high rates of diabetes within this population, leading to extensive studies aimed at understanding the factors contributing to this health issue. (Schulz & Chaudhari, 2015)

One of the key factors under investigation is the impact of the number of pregnancies on the likelihood of adult women, specifically those aged 21 and above, developing diabetes. Previous research studies, including the one by Akshay Dattatray Khare, have suggested a potential link between the number of pregnancies a woman experiences and her susceptibility to diabetes (Khare, 2022). Pregnancy induces significant hormone changes in a woman’s body, which can influence insulin sensitivity. For Pima Indian women, exploring this relationship is of particular interest due to the higher prevalence of diabetes in their community. 

To investigate this relationship, researchers often rely on comprehensive datasets that include information on the number of pregnancies, diabetes status, genetic factors, lifestyle choices and other relevant variables. Analyzing such data sets, like the one available on Kaggle, provides a valuable opportunity to explore the correlation between the number of pregnancies and the likelihood of diabetes in Pima Indian women.(Khare, 2022) By conducting a rigorous analysis of this data, researchers aim to contribute meaningful insights to the field of public health, potentially paving the way for more effective diabetes prevention tailored specifically to these Pima Indian women. (Khare, 2022)

Researchers have found a link between Gestational diabetes and pregnancies. (Gestational diabetes and pregnancy, 2022) Other researchers, including Eoin Noctor, have shown that they also focus on glucose levels as well. He also talked about the heterogeneity of risk factors, including pre-pregnancy BMI, gestational age at diagnosis, and ethnicity, contributing to the understanding of the progression to type 2 diabetes post-GDM (Noctor, 2015).

First, we need to split the raw dataset `diabetes` into three datasets: `train`, `test`, `validation` with the ratio $6:2:2$. 

```python
# function for spliting dataset into train, test, validation set with specific ratio
def train_test_valid_split(dataset, test_size=None, valid_size=None):
    y = dataset['Outcome']
    X = dataset.drop('Outcome', axis=1)
    x_trainvalid, x_test, y_trainvalid, y_test = train_test_split(X, y, test_size=test_size, random_state=444)
    x_train, x_valid, y_train, y_valid = train_test_split(x_trainvalid, y_trainvalid, 
                                                        test_size=valid_size/(1-test_size), random_state=444)
    
    print("training set size: ", x_train.shape, y_train.shape, round(x_train.shape[0]/dataset.shape[0], 2))
    print("testing set size: ", x_test.shape, y_test.shape, round(x_test.shape[0]/dataset.shape[0], 2))
    print("validation set size: ", x_valid.shape, y_valid.shape, round(x_valid.shape[0]/dataset.shape[0], 2))

    return x_train, x_test, x_valid, y_train, y_test, y_valid 
```

```python
# split Diabetes into train, test, validation set with the ratio: 6:2:2
x_train_diabetes, x_test_diabetes, x_valid_diabetes, y_train_diabetes, y_test_diabetes, y_valid_diabetes = train_test_valid_split(diabetes, test_size=0.2, valid_size=0.2)
```

In **Data** section, we mentioned a list of metrics for different explanatory variables. In the next part, we will discretize data(from each of three sets `train`,`test`,`validation`) according to this list of metrics. For those are equal to or above the metric, we set the value to be 1. Otherwise, setting the value to be 0. 

In EDA section, this part will help us to calculate the proportion of each explanatory variables associated to different outcomes of the response variable `Outcome`.

```python
# function to discretize data for Diabetes
def discrete_diabetes(dataset):
    discrete = pd.DataFrame(columns=['Pregnancies_Over2','Glucose_Over126','BloodPressure_Over90',
                                        'BMI_Over25','Age_Over40','Outcome'])

    discrete['Pregnancies_Over2'] =  np.where(dataset['Pregnancies']>=2, 1, 0)
    discrete['Glucose_Over126'] = np.where(dataset['Glucose']>=126, 1, 0)
    discrete['BloodPressure_Over90'] = np.where(dataset['BloodPressure']>=90, 1, 0)
    discrete['BMI_Over25'] = np.where(dataset['BMI']>=25, 1, 0)
    discrete['Age_Over40'] = np.where(dataset['Age']>=40, 1, 0)
    discrete['Outcome'] = dataset['Outcome']

    return discrete
```

```python
# view discretized data - Diabetes training set 
train_discDiabetes = discrete_diabetes(train_diabetes)
print("training set size - discrete: ", train_discDiabetes.shape)
train_discDiabetes.head(7)
```

```python
# view discretized data - Diabetes testing set
test_discDiabetes = discrete_diabetes(test_diabetes)
print("testing set size - discrete: ", test_discDiabetes.shape)
test_discDiabetes.head(7)
```

```python
# view discretized data - Diabetes validation set
valid_discDiabetes = discrete_diabetes(valid_diabetes)
print("validation set size - discrete: ", valid_discDiabetes.shape)
valid_discDiabetes.head(7)
```

Here, we create a new function `discrete_GestationalD` for discretizing data of all three sets. The reason of why no longer using the function `discrete_diabetes` here is that: two datasets `[set]_diabetes` and `[set]_GestationalD` contains different variables, the number of variables is also different. According to the list of metrics for `GestationalD` in **Data** section, it contains more metrics. Applying `discrete_diabetes` here will cause problem and will return error. 

```python
# function to discretize data for Gestational Diabetes
def discrete_GestationalD(dataset):

    discrete = pd.DataFrame(columns=['Age_Over40','Pregnancies_Over2','GestationHistory_Over1',
                                        'BMI_Over25', 'FamilyHistory','DiaBP_Over90',
                                        'Hemoglobin_Over13','Prediabetes','Outcome'])

    discrete['Age_Over40'] = np.where(dataset['Age']>=40, 1, 0)
    discrete['Pregnancies_Over2'] =  np.where(dataset['Pregnancies']>=2, 1, 0)
    discrete['GestationHistory_Over1'] = np.where(dataset['GestationHistory']>=1, 1, 0)
    discrete['BMI_Over25'] = np.where(dataset['BMI']>=25, 1, 0)
    discrete['FamilyHistory'] = dataset['FamilyHistory']
    discrete['DiaBP_Over90'] = np.where(dataset['DiaBP']>=90, 1, 0)
    discrete['Hemoglobin_Over13'] = np.where(dataset['Hemoglobin']>=13, 1, 0)
    discrete['Prediabetes'] = dataset['Prediabetes']
    discrete['Outcome'] = dataset['Outcome']

    return discrete
```

```python
# function for visualizing the relationship of each Exploratory variable and Outcome
def VarOutcome_relationship(dataset):
    i = 1
    plt.figure(figsize=(15, 15))
    for variable in dataset.columns:
        if variable != 'Outcome':
            plt.subplot(math.ceil((dataset.shape[1]-1)/2),2,i)
            data_0 = dataset[dataset['Outcome']==0][variable]
            data_1 = dataset[dataset['Outcome']==1][variable]
            pdf_0 = stats.norm.pdf(data_0, loc=data_0.mean(), scale=data_0.std())
            pdf_1 = stats.norm.pdf(data_1, loc=data_1.mean(), scale=data_1.std())
            sns.lineplot(x=data_0, y=pdf_0, 
                label='Outcome=0, max_pdf='+str(round(pdf_0.max(),4)))
            sns.lineplot(x=data_1, y=pdf_1, 
                label='Outcome=1, max_pdf='+str(round(pdf_1.max(),4)))
            plt.ylabel('conditional probability density (pdf)')
            i += 1

    plt.tight_layout()
    plt.show()
    return
```

```python
# function for creating the proportion table for discretized data and outcome
def proportion_table(variable, dataset):
    if variable != 'Outcome':
        variable_counts = dataset.value_counts(variable)
        prop_df = dataset[[variable, 'Outcome']]
        prop_df = prop_df.groupby([variable, "Outcome"], group_keys=True).value_counts()/variable_counts
        prop_df = prop_df.reset_index()
        prop_df.columns = [variable, 'Outcome', 'proportion']
        return prop_df
```

```python
# Barplot for the corresponding proportion table
def proportion_barplot(dataset):
    i = 1
    plt.figure(figsize=(15, 15))
    for variable in dataset.columns:
        if variable != 'Outcome':
            prop_table = proportion_table(variable, dataset)
            plt.subplot(math.ceil((dataset.shape[1]-1)/2),2,i)
            sns.barplot(prop_table, x="Outcome", y="proportion", hue=variable)
            i += 1
    
    plt.tight_layout()
    plt.show()
    return
```

# Research Question

How does the number of pregnancies of an adult woman(at least 21 years old) of Pima Indian heritage affect their probability of getting diabetes(Type-II)?


## Background and Prior Work

The Pima Indians, also known as the Akimel O’odham, are a Native American tribe predominantly residing in the Salt River Pima–Maricopa Indian Community in Arizona, USA. This community has received attention in the realm of public health due to a high prevalence of diabetes, both Type-I and Type-II, among its members. Researchers have long been intrigued by the unusually high rates of diabetes within this population, leading to extensive studies aimed at understanding the factors contributing to this health issue. (Schulz & Chaudhari, 2015)

One of the key factors under investigation is the impact of the number of pregnancies on the likelihood of adult women, specifically those aged 21 and above, developing diabetes. Previous research studies, including the one by Akshay Dattatray Khare, have suggested a potential link between the number of pregnancies a woman experiences and her susceptibility to diabetes (Khare, 2022). Pregnancy induces significant hormone changes in a woman’s body, which can influence insulin sensitivity. For Pima Indian women, exploring this relationship is of particular interest due to the higher prevalence of diabetes in their community. 

To investigate this relationship, researchers often rely on comprehensive datasets that include information on the number of pregnancies, diabetes status, genetic factors, lifestyle choices and other relevant variables. Analyzing such data sets, like the one available on Kaggle, provides a valuable opportunity to explore the correlation between the number of pregnancies and the likelihood of diabetes in Pima Indian women.(Khare, 2022) By conducting a rigorous analysis of this data, researchers aim to contribute meaningful insights to the field of public health, potentially paving the way for more effective diabetes prevention tailored specifically to these Pima Indian women. (Khare, 2022)

Researchers have found a link between Gestational diabetes and pregnancies. They have studied this link by assessing factors such as glucose levels, insulin resistance, lifestyle choices and BMI during and after pregnancy. (Gestational diabetes and pregnancy, 2022) Other researchers, including Eoin Noctor, have shown that they also focus on glucose levels as well. He also talked about the heterogeneity of risk factors, including pre-pregnancy BMI, gestational age at diagnosis, and ethnicity, contributing to the understanding of the progression to type 2 diabetes post-GDM (Noctor, 2015). 

We plan to use these same variables in our data as some of the researchers we have seen have done, with one response variable and 8 explanatory variables. We decided to put gestational diabetes and type II diabetes in the same discussion because both conditions can impact maternal and fetal health. Researching them together helps in understanding the shared consequences, complications, and potential interventions to improve outcomes for both mothers and infants.

## Dataset #1: Diabetes

```python
## Import library
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
## Load Data - Diabetes
diabetes = pd.read_csv('DataSets/diabetes.csv')
print("shape of diabetes dataset: ", diabetes.shape)
diabetes.head(7)
```

```python
# check if the dataset contains NaN/null value
# if so, drop all observations that contains NaN/Null value
diabetes.isna().sum()
```

```python
#drop columns
diabetes = diabetes.drop(['SkinThickness', 'Insulin', 'DiabetesPedigreeFunction'],axis=1)
```

```python
#change specific column name
diabetes = diabetes.rename(columns={'Pregnancies':'No. of Pregnancy'})
```

```python
#create new column age_over40
diabetes['Age_Over40'] = np.where(diabetes['Age']>=40, 1, 0)
```

```python
#view the updated Diabetes
print("shape of the updated diabetes dataset: ", diabetes.shape)
diabetes.head(7)
```

## Dataset #2: Gestational Diabetes

```python
## Load Data - Gestational Diabetes
GestationalD = pd.read_excel("Datasets/Gestational Diabetic Dat Set.xlsx")
print("shape of Gestational Diabetes: ", GestationalD.shape)
GestationalD.head(7)
```

```python
#drop columns
GestationalD = GestationalD.drop(['Case Number', 'HDL', 'unexplained prenetal loss', 'Large Child or Birth Default', 'PCOS', 'Sys BP', 'OGTT', 'Sedentary Lifestyle'], axis=1)
GestationalD.isna().sum()
```

```python
# drop NaN value
# keep observations with Age >= 21
GestationalD = GestationalD.dropna(subset = ['BMI'])
GestationalD = GestationalD[GestationalD['Age'] >= 21]
GestationalD = GestationalD.reset_index(drop=True)
```
